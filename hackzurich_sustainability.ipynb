```python
data = pd.read_csv('/kaggle/input/sustainability.csv')
data.head()
```

```python
data['sustainability_score'].describe()
```

```python
def simplify_sustainability(data):
    data.sustainability_score = data.sustainability_score.fillna(-0.5)
    bins = (-1, 0, 17, 41, 65, 100)
    group_names= ['Unknown', '1_quartile', '2_quartile', '3_quartile', '4_quartile']
    categories = pd.cut(data.sustainability_score, bins, labels = group_names)
    data.sustainability_score = categories
    return data
```

# Lets make this problem a classification problem

```python
simplify_sustainability(data)
```

```python
features = data
labels = data.sustainability_score.values
```

```python
data = data.drop(columns=['company','sustainability_score'])
 #df.drop(columns =['C', 'D']) 
data = data.drop(data.columns[0], axis=1)
```

```python
data.head()
```

```python
# encoding the labels
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
le = encoder.fit(labels)
labels = le.transform(labels)

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data, labels, test_size =0.20,random_state = 2)
```

```python
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
#from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer,accuracy_score
import numpy as np

acc_score = make_scorer(accuracy_score)

#Model 4 - XGBoost Classifier
#def xgboost(X_train,X_test,y_train,y_test,useTrainCV=True,cv_folds=5,early_stopping_rounds = 50):
model = XGBClassifier()
model.fit(X_train, y_train)
  
        
# make predictions for test data
y_pred = model.predict(X_test)
#predictions = [round(value) for value in y_pred]
# evaluate predictions
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
